###### solution--> we will assume middle stage 'doggo' for all rows have 'floofer' Stage as a floofer is also a doggo

###### define
###### 3-df_image_predictions --> prediction variable not in one column
###### 4 -df_image_predictions --> dog variable not in one column
###### 5 -df_image_predictions --> confindence variable not in one column
###### take best pediction 

###### code

```python
def best_prediction(confidences,predictions):
    return (predictions[confidences.index(max(confidences))],max(confidences))

prediction=[]
is_dog=[]
confindence=[]
for index,row in cleaned_df_image_predictions.iterrows():
    _is_dog=row.prediction_1_is_dog|row.prediction_2_is_dog|row.prediction_3_is_dog
    is_dog.append(_is_dog)
    if(_is_dog):
        best_prediction_and_confindence=best_prediction([row.prediction_1_confindence,row.prediction_2_confindence,row.prediction_3_confindence],[row.prediction_1,row.prediction_2,row.prediction_3])
        prediction.append(best_prediction_and_confindence[0])
        confindence.append(best_prediction_and_confindence[1])
        print('appended')
    else:
        prediction.append('no_prediction')
        confindence.append('no_confindence')
        print('failed')
        
cleaned_df_image_predictions['prediction']=prediction
cleaned_df_image_predictions['is_dog']=is_dog
cleaned_df_image_predictions['confindence']=confindence
```

```python
isDog_rating_means=df.groupby('is_dog_int')['rating'].mean().sort_values(ascending=False).nlargest(10)
ax=isDog_rating_means.plot(kind='bar',figsize=(10,10))
plt.ylabel('Mean Rating')
plt.xlabel('is Dog')
plt.title('Fig.5 Tweets with image likely have higher Rating',fontsize=20, y=-0.15)
ax.set_xticklabels(['True','Fales'], rotation=0,fontsize=16)
for i in ax.patches:
    ax.text(i.get_x()+.15, i.get_height()+.05,
            str(round(i.get_height(),2)), fontsize=20)
```

```python
retweet_mean=df.set_index('time_stamp').groupby(pd.Grouper(freq='M'))['retweet_count'].mean()
favorite_mean=df.set_index('time_stamp').groupby(pd.Grouper(freq='M'))['favorite_count'].mean()
fig, ax = plt.subplots(figsize=(10,10))
retweet_mean.plot(label='retweet_mean')
favorite_mean.plot(label='favorite_mean')
leg = ax.legend();
plt.xlabel('Time Stamp')
plt.ylabel('Count')
plt.title('Fig.6 Interaction over time',fontsize=20, y=-0.15)
```

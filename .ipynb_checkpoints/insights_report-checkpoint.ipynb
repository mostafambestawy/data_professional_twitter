###### test

```python
values=cleaned_df_image_predictions.prediction_1.value_counts()
i=0
while(True):
    if i+50 < len(values):
        print(values[i:i+50])
        i=i+50
    else:
        break
```

###### define
###### 3-df_image_predictions --> prediction variable not in one column
###### 4 -df_image_predictions --> dog variable not in one column
###### 5 -df_image_predictions --> confindence variable not in one column
###### take best pediction 

###### code

```python
def best_prediction(confidences,predictions):
    return (predictions[confidences.index(max(confidences))],max(confidences))

prediction=[]
is_dog=[]
confindence=[]
for index,row in cleaned_df_image_predictions.iterrows():
    _is_dog=row.prediction_1_is_dog|row.prediction_2_is_dog|row.prediction_3_is_dog
    is_dog.append(_is_dog)
    if(_is_dog):
        best_prediction_and_confindence=best_prediction([row.prediction_1_confindence,row.prediction_2_confindence,row.prediction_3_confindence],[row.prediction_1,row.prediction_2,row.prediction_3])
        prediction.append(best_prediction_and_confindence[0])
        confindence.append(best_prediction_and_confindence[1])
        print('appended')
    else:
        prediction.append('no_prediction')
        confindence.append('no_confindence')
        print('failed')
        
cleaned_df_image_predictions['prediction']=prediction
cleaned_df_image_predictions['is_dog']=is_dog
cleaned_df_image_predictions['confindence']=confindence
```
